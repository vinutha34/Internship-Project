```python
import pandas as pd
import folium
df = pd.read_csv("Dataset.csv")
```

```python
map_rest = folium.Map(
    location=[df['Latitude'].mean(), df['Longitude'].mean()],
    zoom_start=5
)

for _, row in df.sample(500).iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=2
    ).add_to(map_rest)

map_rest
```
